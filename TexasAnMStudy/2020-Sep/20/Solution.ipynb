# Userful Formula
## Periodicity
$sin(\theta + 2\pi) = sin(\theta)$  
Therefore, $sin(\theta + 2n\pi) = sin(\theta)$ when n is any integer

Likewise,  
$cos(\theta + 2n\pi) = cos(\theta)$  
Especially,  
$tan(\theta + n\pi) = tan(\theta)$  

## Transpose
$sin(\theta + \frac{\pi}{2}) = cos(\theta)$  
$sin(\theta - \frac{\pi}{2}) = -cos(\theta)$  
$cos(\theta + \frac{\pi}{2}) = -sin(\theta)$  
$cos(\theta - \frac{\pi}{2}) = -sin(\theta)$  
$tan(\theta + \frac{\pi}{2}) = -\frac{1}{tan(\theta)} = -atan(\theta)$  
$tan(\theta - \frac{\pi}{2}) = \frac{1}{tan(\theta)} = atan(\theta)$  

## Symmetry
$sin(-\theta) = -sin(\theta)$ (Odd function)  
$cos(-\theta) = cos(\theta)$ (Even function)  
$tan(-\theta) = -tan(\theta)$ (Odd function)


# Visualization using python

```python
import numpy as np; import scipy as sp; import matplotlib.pyplot as plt
```

```python
theta = [0.01*i for i in range(-200, 201)]
y_sin = [np.sin(j*np.pi) for j in theta]
plt.plot(theta, y_sin)
```

```python
y_cos = [np.cos(j*np.pi) for j in theta]
plt.plot(theta, y_cos)
```

```python
y_tan = [np.tan(j*np.pi) for j in theta]
plt.plot(theta, y_tan)
```
